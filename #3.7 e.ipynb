# Question 3.7

e) Is multicollinearity a potential problem in this model? 


*** To find if multicollinearity is an issue in this model, there are two ways to figure out.*** 


# First Way

One of them is ScatterMatrix plot or Correlation Matrix; if there is strong linear relationship between two regressor variables, there exists a multicollinearity.

```r
data = read.csv("B4.csv")
```

```r
matrix = cor(data)
options(repr.plot.width=5, repr.plot.height=5)
corrplot.mixed(matrix, upper="pie", lower="number") 
```

***Conclusion***

Following regressor variables are strongly correlated:-

x6 & x7 : r = 0.87

x1 & x4 : r = 0.73

x2 & x4 : r = 0.73

x1 & x3 : r = 0.69

There exists a multicollinearity in this model.

# Second Way

Second way is to calculate VIF(Variation Inflation Factor). It is reccommended that if VIF value > 10, there exists a multicollinearity.

<img src= https://lh3.googleusercontent.com/3quzbxMvc4SVETw7VcZ37oOc1bbLxhWJVncMILlg-Gf-7ppcnXTqjTeNZOoGJ_NrhkheQIuIF8ikmk-GiOKahau0-JyFRDGvR6q2E1QwYCmh0HhYMharaBdFCBfnfittlK1anjvh />


```r
y= data$y
reg <- lm(y~x1 + x2 + x3 + x4 + x5 + x6 + x7 + x8 + x9, data = data)
summary(reg)
```

```r
car::vif(reg)
```

VIF of $x_6$ is 11.7 which is greater than 10 and VIF of $x_7$ is 9.7 which is significat as it closer to 10. This suggests that there exists multicollinearity in this model. 

# Overall Conclusion 

***Both ways shows that this model has multicollinearity issue because of strong linear relationship between $x_6 and x_7$ variables.***
